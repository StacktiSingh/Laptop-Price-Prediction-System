```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get(f"https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page=1")
response.status_code
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-IN,en;q=0.9"
}
page_count = 1
while True:
    URL = f"https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={page_count}"
    response = requests.get(URL, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    laptops = soup.select(".QSCKDh.dLgFEE .lvJbLV.col-12-12")
    if not laptops:
        print("NO valid pages anymore")
        break
    with open(f'laptop_data_{page_count}', 'w', encoding='utf-8') as file:
        file.write(response.text)
        print("Downloaded data from page", page_count)
    page_count += 1
    
```

```python
from bs4 import BeautifulSoup

all_laptops = []

for i in range(1, 42):
    with open(f"laptop_data_{i}", encoding="utf-8") as data:
        soup = BeautifulSoup(data, "html.parser")
        cards = soup.select("div.QSCKDh.dLgFEE")
        for card in cards:
            specs = card.select_one("ul.HwRTzP")
            if not specs:
                continue
            specs_items = specs.find_all("li")
            specs_text = [li.get_text(strip=True) for li in specs_items]
            
            record = {
            "brand": None,
            "price": None,
            "processor": None,
            "ram_gb": None,
            "storage_gb": None,
            "storage_type": None,
            "screen_size": None,
            "os": None
        }
            title = card.select_one("div.RG5Slk")
            if title:
                record['brand'] = title.get_text(strip=True).split()[0]
            price_tag = card.select_one("div.hZ3P6w.DeU9vF")
            if price_tag:
                record['price'] = int(price_tag.get_text(strip=True).replace("₹", "").replace(",", ""))
            for item in specs_text:
                text = item.lower()
        
                if "ram" in text:
                    record["ram_gb"] = int(text.split()[0])
        
                elif "ssd" in text or "hdd" in text:
                    record["storage_gb"] = int(text.split()[0])
                    record["storage_type"] = "SSD" if "ssd" in text else "HDD"
        
                elif "processor" in text or "core" in text or "ryzen" in text:
                    record["processor"] = item
        
                elif "inch" in text:
                    record["screen_size"] = float(text.split()[0])
        
                elif "windows" in text or "dos" in text or "mac" in text:
                    record["os"] = item
        
        all_laptops.append(record)
```
